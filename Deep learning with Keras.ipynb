```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot  as plt
```

```python
x= np.linspace (-1,1,1001)
y= 2 * x+ np.random.randn(*x.shape) * 0.33

plt.scatter(x,y)
```

```python
def createline(x,w=0,b=0):
    return w*x+b
```

```python
plt.scatter(x,y)
w=2.0087843
b= -0.0057927
predict = createline(x,w,b)

plt.plot(x,predict, c='r')
plt.show()
```

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(1,activation='relu', input_dim=1))
```

```python
model.compile(optimizer='rmsprop' , loss= 'mse' , metrics=['mae','accuracy', 'mse'])
```

```python
 model.fit(x,y, epochs=100)
```

```python
model.summary()
```

# Softmax Function Example

```python
def softmax(x):
    prob = []
    for i in x:
        b=i/np.sum(x)
        prob.append(b)
    return prob
```

```python
data =np.array([1,3,5,6,7])
softmax(data)
```

```python
np.sum(softmax(data))
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1. / (1. +np.exp(-x))
```

```python
test = np.linspace(-10,10,101)
y1=sigmoid(test)
y1
```

```python
plt.plot(y1)
plt.show()
```

# RELU

```python
def relu(x):
    x[x<=0] = 0
    x[x>0] = 1
    return x
```

```python
y1  =  relu(test)
y1
```

```python
plt.plot(y1)
plt.show()
```

```python
from keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
import numpy as  np
from keras import models
from keras import layers
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
```

# Network Compile

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python

train_images  = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
train_images.size
```

```python
from keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```
